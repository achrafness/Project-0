```python
data_set = pd.read_csv("/kaggle/input/data-set-raftel/challenge_train.csv")
data_set 
```

```python
data_set.isna().sum()

data_set = data_set.fillna(value=data_set["Age"].mean())
data_set = data_set.fillna(value=data_set["journey Sub Category Cipher"].mode())
data_set = data_set.fillna(value=data_set["Treasure Technology Type"].mode())
data_set = data_set.fillna(value=data_set["Skill Level"].mode())
data_set = data_set.fillna(value=data_set["Pirate Level"].mode())
data_set = data_set.fillna(value=data_set["Pirate Still Working"].mode())
data_set = data_set.fillna(value=data_set["University Degree Score"].mean())
data_set = data_set.fillna(value=data_set["Pirate Job Type"].mode())
data_set = data_set.fillna(value=data_set["Pirate Employment Status"].mode())
data_set = data_set.fillna(value=data_set["University Degree Score System"].mode())
data_set = data_set.fillna(value=data_set["University Degree Score"].mode())
data_set['journey End Date'] = pd.to_datetime(data_set['journey End Date'], format='%Y-%m-%d')
data_set['journey Start Date'] = pd.to_datetime(data_set['journey Start Date'], format='%Y-%m-%d')

data_set
```

```python
 from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# Split data into features and target variable
features = ['Age', 'journey Sub Category Cipher', 'Treasure Technology Type', 'Skill Level', 
            'Pirate Level', 'Pirate Still Working', 'University Degree Score', 
            'Pirate Job Type', 'Pirate Employment Status', 'University Degree Score System']
target = 'Arriving to Raftel'

non_numeric_cols = data_set.select_dtypes(exclude=['float', 'int']).columns
data_set = data_set.drop(columns=non_numeric_cols)

# Split data into features and target variable
features = data_set.columns.tolist()
features.remove('Arriving to Raftel')

X_train = data_set[features]
y_train = data_set[target]



# Train a Random Forest classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

non_numeric_cols = data_set.select_dtypes(exclude=['float', 'int']).columns
data_set = data_set.drop(columns=non_numeric_cols)

# Split data into features and target variable
features = data_set.columns.tolist()
features.remove('Arriving to Raftel')

X_train = data_set[features]
y_train = data_set[target]


test_set = pd.read_csv("/kaggle/input/data-set-raftel/challenge_test.csv")

test_set.isna().sum()

test_set = test_set.fillna(value=test_set["Age"].mean())
test_set = test_set.fillna(value=test_set["journey Sub Category Cipher"].mode())
test_set = test_set.fillna(value=test_set["Treasure Technology Type"].mode())
test_set = test_set.fillna(value=test_set["Skill Level"].mode())
test_set = test_set.fillna(value=test_set["Pirate Level"].mode())
test_set = test_set.fillna(value=test_set["Pirate Still Working"].mode())
test_set = test_set.fillna(value=test_set["University Degree Score"].mean())
test_set = test_set.fillna(value=test_set["Pirate Job Type"].mode())
test_set = test_set.fillna(value=test_set["Pirate Employment Status"].mode())
test_set = test_set.fillna(value=test_set["University Degree Score System"].mode())
test_set = test_set.fillna(value=test_set["University Degree Score"].mode())
test_set['journey End Date'] = pd.to_datetime(test_set['journey End Date'], format='%Y-%m-%d')
test_set['journey Start Date'] = pd.to_datetime(test_set['journey Start Date'], format='%Y-%m-%d')

X_test = test_set[features]
predictions = model.predict(X_test)

# Format result
result = pd.DataFrame({'ID': test_set.index + 1 , 'Arriving to Raftel': predictions})
# Save the results to a CSV file
result.to_csv("result.csv", index=False)
result
```
